```python
def parse_trainer(x):

    """
    Converts trainer field to:
      - trainer_list: a list of trainer IDs
      - trainer_primary: first trainer or 'Unknown'
      - n_trainers: number of trainers
    Handles cases where x is:
      - already a string like 'TRA_xxx'
      - a string representation of a list like "['TRA_xxx']"
      - missing (NaN)
    """

    if pd.isna(x):
        return [], "Unknown", 0

    if isinstance(x, str):
        x = x.strip()

        # If list-like string
        if x.startswith("[") and x.endswith("]"):
            try:
                trainers = ast.literal_eval(x)
            except:
                trainers = []
        else:
            trainers = [x]
    else:
        trainers = [str(x)]

    trainers = [t.strip() for t in trainers if isinstance(t, str) and t.strip() != ""]

    trainer_primary = trainers[0] if len(trainers) > 0 else "Unknown"
    n_trainers = len(trainers)

    return trainers, trainer_primary, n_trainers
```

```python
# Apply parser to Train/Test/Prior

def add_trainer_features(df):
    trainer_lists = []
    trainer_primary = []
    n_trainers = []

    for x in df["trainer"]:
        t_list, t_primary, n_t = parse_trainer(x)
        trainer_lists.append(t_list)
        trainer_primary.append(t_primary)
        n_trainers.append(n_t)

    df = df.copy()
    df["trainer_list"] = trainer_lists
    df["trainer_primary"] = trainer_primary
    df["n_trainers"] = n_trainers
    return df

train = add_trainer_features(train)
test = add_trainer_features(test)
prior = add_trainer_features(prior)

train[['trainer', 'trainer_primary', 'n_trainers']].head(10)
                            
```

```python
# Parse topics_list

def parse_topics(x):
    """"
    converts topic_list field to a clean list of topics.
    Handles:
        -string representation of list
        -actual list
        -missing values
    """

    # Handle missing at the top level
    if pd.isna(x):
        return []

    # If it's already a list
    if isinstance(x, list):
        topics = x

    # If it's a string (most common case)
    elif isinstance(x, str):
        s = x.strip()
        if s.startswith("[") and s.endswith("]"):
            try:
                topics = ast.literal_eval(s)
            except Exception:
                topics = []
        else:
            topics = [s]

    else:
        topics = [str(x)]

    # IMPORTANT: Flatten nested lists
    flat_topics = []
    for item in topics:
        if isinstance(item, list):
            flat_topics.extend(item)
        else:
            flat_topics.append(item)

    # Clean each topic safely
    clean = []
    for t in flat_topics:
        if isinstance(t, str):
            t = t.strip()
            if t != "":
                clean.append(t)

    return clean

```

```python
# Add topic list + basic count features

def add_topic_features(df):
    df = df.copy()
    df["topics_parsed"] = df["topics_list"].apply(parse_topics)

    df["n_topics"] = df["topics_parsed"].apply(len)
    df["n_unique_topics"] = df["topics_parsed"].apply(lambda x: len(set(x)))
    df["topic_diversity"] = df["n_unique_topics"] / df["n_topics"].replace(0, np.nan)
    df["topic_diversity"] = df["topic_diversity"].fillna(0.0)

    return df

train = add_topic_features(train)
test  = add_topic_features(test)
prior = add_topic_features(prior)

train[["topics_list", "topics_parsed", "n_topics", "n_unique_topics", "topic_diversity"]].head(20)
```

```python
# Create top-k multi hot topic features

from collections import Counter

def get_top_topics (train_df, prior_df, k = 50):
    all_topics = []
    all_topics += train_df["topics_parsed"].sum() #sums lists together
    all_topics += prior_df["topics_parsed"].sum()

    topic_counts = Counter(all_topics)
    top_topics = [t for t, c in topic_counts.most_common(k)]
    return top_topics, topic_counts

TOP_K = 50
top_topics, topic_counts = get_top_topics(train, prior, k = TOP_K)

print("Top 10 Topics:")
print(top_topics[:10])
```

```python
# Add multi-hot columns to Train/test/prior

def add_multihot_topics(df, top_topics):
    df = df.copy()
    topic_set_list = df["topics_parsed"].apply(set)
    
    for t in top_topics:
        df[f"topic__{t}"] = topic_set_list.apply(lambda s: 1 if t in s else 0)

    return df

train = add_multihot_topics(train, top_topics)
prior = add_multihot_topics(prior, top_topics)
test = add_multihot_topics(test, top_topics)

# quick check
topic_cols = [c for c in train.columns if c.startswith("topic__")]
print("Number of topic multi-hot columns:", len(topic_cols))
train[topic_cols].head(20)
```

**The leakage rule (IMPORTANT)**

We must ensure:

When creating features for a row at date D, we only use prior rows with date < D. We will enforce this with merge_asof(..., allow_exact_matches=False) which prevents using same-day rows.

```python
# Sort data (required for merge_asof)

train = train.sort_values(["farmer_name", "training_day"]).reset_index(drop = True)
test = test.sort_values(["farmer_name", "training_day"]).reset_index(drop = True)
prior = prior.sort_values(["farmer_name", "training_day"]).reset_index(drop = True)
```

```python
# Build “cumulative history” features inside Prior

# compute farmer history over time inside prior, then shift by 1 so an event doesn’t “see itself”.

prior_feat = prior.copy()

# 1) count event per farmer over time (cumulative)
prior_feat["prior_n_events_cum"] = prior_feat.groupby("farmer_name").cumcount() + 1

# 2) first training day per farmer
prior_feat["prior_first_day"] = prior_feat.groupby("farmer_name")["training_day"].transform("min")

# 3) previous training day per farmer
prior_feat["prior_prev_day"] = prior_feat.groupby("farmer_name")["training_day"].shift(1)

# 4) days since first or last at each prior row (inside prior timeline)
prior_feat["prior_days_since_first"] = (prior_feat["training_day"] - prior_feat["prior_first_day"]).dt.days
prior_feat["prior_days_since_last"] = (prior_feat["training_day"] - prior_feat["prior_prev_day"]).dt.days

prior_feat[["farmer_name","training_day","prior_first_day","prior_prev_day","prior_days_since_first","prior_days_since_last"]].head(20)
```

```python
# Shift cumulative features to make them safe. We want features at time D to reflect history before D, not including D.

prior_feat["prior_n_events_before"] = prior_feat.groupby("farmer_name")["prior_n_events_cum"].shift(1)

# For "days since first" and "days since last", we want the values as of the previous event,
# but we’ll compute final versions after merging into train/test.
```

```python
# Check if columns exist in prior

adopt_cols = ["adopted_within_07_days", "adopted_within_90_days", "adopted_within_120_days"]
print("Adopt columns in prior:", [c for c in adopt_cols if c in prior_feat.columns])
```

```python
# build cumulative sums and rates

for col in adopt_cols:
    if col in prior_feat.columns:
        # cumulative sum of adoptions
        prior_feat[f"{col}_cum_sum"] = prior_feat.groupby("farmer_name")[col].cumsum()
        # shift so it's "before current event"
        prior_feat[f"{col}_cum_sum_before"] = prior_feat.groupby("farmer_name")[f"{col}_cum_sum"].shift(1)

# adoption rates before current event = cum_adoptions_before / n_events_before
for col in adopt_cols:
    if col in prior_feat.columns:
        prior_feat[f"prior_rate_{col.split('_within_')[1]}"] = (
            prior_feat[f"{col}_cum_sum_before"] / prior_feat["prior_n_events_before"]
        )
```

```python
# Merge Prior history into Train/Test (leakage-safe). use merge_asof to attach each train row to the farmer’s latest prior row strictly before that date.

# Choose which prior features to carry forward

keep_cols = [
    "farmer_name", "training_day",
    "prior_n_events_before",
    "prior_first_day",
    "prior_prev_day"
]

# Include rates if they exist
rate_cols = [c for c in prior_feat.columns if c.startswith("prior_rate_")]
keep_cols += rate_cols

prior_for_merge = prior_feat[keep_cols].sort_values(["farmer_name", "training_day"])
```

```python
# Ensure datetime
train["training_day"] = pd.to_datetime(train["training_day"])
prior_for_merge["training_day"] = pd.to_datetime(prior_for_merge["training_day"])

# Correct sorting order for asof
train_left = train.sort_values(["training_day", "farmer_name"]).reset_index(drop=True)
prior_right = prior_for_merge.sort_values(["training_day", "farmer_name"]).reset_index(drop=True)

# merge for train
train_merged = pd.merge_asof(
    train_left,
    prior_right,
    by="farmer_name",
    on="training_day",
    direction="backward",
    allow_exact_matches=False
)

```

```python
# merge for test

test_left = test.sort_values(["training_day", "farmer_name"]).reset_index(drop=True)
test_left["training_day"] = pd.to_datetime(test_left["training_day"]).reset_index(drop = True)

test_merged = pd.merge_asof(
    test_left,
    prior_right,
    by="farmer_name",
    on="training_day",
    direction="backward",
    allow_exact_matches=False
)
```

```python
# Create final history features after merge

# Compute:
   # history exists?
   # days since last training
   # days since first training
   # fill missing values (farmers with no prior history)

def finalize_history_features(df):

    df = df.copy()

    # if no prior match, these will be NaN
    df["has_prior_history"] = df["prior_n_events_before"].notna().astype(int)

    # replace mssing history with 0 training before
    df["prior_n_events_before"] = df["prior_n_events_before"].fillna(0)

    # days since first/last
    df["prior_days_since_first"] = (df["training_day"] - df["prior_first_day"]).dt.days
    df["prior_days_since_last"] = (df["training_day"] - df["prior_prev_day"]).dt.days

    # fill missing with -1 meaning "no history"
    df["prior_days_since_first"] = df["prior_days_since_first"].fillna(-1)
    df["prior_days_since_last"] = df["prior_days_since_last"].fillna(-1)

    # fill missing adoption rates with 0 (no evidence of adoption)
    rate_cols = [c for c in df.columns if c.startswith("prior_rate_")]
    for c in rate_cols:
        df[c] = df[c].fillna(0)

    return df

train_merged = finalize_history_features(train_merged)
test_merged = finalize_history_features(test_merged)


train_merged[["farmer_name","training_day","has_prior_history","prior_n_events_before","prior_days_since_last"]].head()
```

use topic multi-hot columns to compute:

```python
# Get the topic columns list

topic_cols = [c for c in train.columns if c.startswith("topic_")]
print("Topic names: ", topic_cols)
print("Number of topic columns: ", len(topic_cols))
```

```python
# Create cumulative topic counts per farmer in prior
# treat each topic column as 0/1 per prior event, then cum-sum over time.

prior_topics = prior[["farmer_name", "training_day"] + topic_cols].copy()
prior_topics = prior_topics.sort_values(["farmer_name","training_day"])

# cumulative count per topic
for c in topic_cols:
    prior_topics[c] = prior_topics.groupby("farmer_name")[c].cumsum()

# shift so it's "before current event"
prior_topics_shift = prior_topics.copy()
prior_topics_shift[topic_cols] = prior_topics_shift.groupby("farmer_name")[topic_cols].shift(1)

prior_topics_shift = prior_topics_shift.sort_values(["farmer_name","training_day"])

```

```python
# Build a renamed version of prior_topics_shift to avoid all collisions forever.

# Identify topic columns (same ones created earlier)
topic_cols = [c for c in train.columns if c.startswith("topic__")]

# Create a copy of prior_topics_shift and rename its topic columns
prior_topics_shift_renamed = prior_topics_shift.copy()
prior_topics_shift_renamed = prior_topics_shift_renamed.rename(
    columns={c: f"priorcount__{c}" for c in topic_cols}
)

# Sort properly for merge_asof
prior_topics_shift_renamed = prior_topics_shift_renamed.sort_values(
    ["training_day", "farmer_name"]
).reset_index(drop=True)
```

```python
# Remove old priorcount columns if they already exist

# Drop any previous priorcount__topic__ columns if they exist in train_merged/test_merged
old_priorcount_cols_train = [c for c in train_merged.columns if c.startswith("priorcount__topic__")]
old_priorcount_cols_test  = [c for c in test_merged.columns if c.startswith("priorcount__topic__")]

train_merged = train_merged.drop(columns=old_priorcount_cols_train, errors="ignore")
test_merged  = test_merged.drop(columns=old_priorcount_cols_test, errors="ignore")
```

```python
# Merge WITHOUT suffixes (no collisions)

# Sort left
train_merged_sorted = train_merged.sort_values(["training_day", "farmer_name"]).reset_index(drop=True)
test_merged_sorted  = test_merged.sort_values(["training_day", "farmer_name"]).reset_index(drop=True)

# Merge
train_merged = pd.merge_asof(
    train_merged_sorted,
    prior_topics_shift_renamed[["training_day","farmer_name"] + [f"priorcount__{c}" for c in topic_cols]],
    by="farmer_name",
    on="training_day",
    direction="backward",
    allow_exact_matches=False
)

test_merged = pd.merge_asof(
    test_merged_sorted,
    prior_topics_shift_renamed[["training_day","farmer_name"] + [f"priorcount__{c}" for c in topic_cols]],
    by="farmer_name",
    on="training_day",
    direction="backward",
    allow_exact_matches=False
)
```

```python
# Compute overlap features
# We want overlap only for topics present today

def add_topic_overlap_features(df, topic_cols):
    df = df.copy()

    today = df[topic_cols].fillna(0).astype(int)
    prior_seen = (df[[f"priorcount__{c}" for c in topic_cols]].fillna(0) > 0).astype(int)

    overlap = (today * prior_seen).sum(axis=1)
    n_today = today.sum(axis=1)

    df["prior_topics_seen_count"] = overlap
    df["prior_topics_seen_ratio"] = (overlap / n_today.replace(0, np.nan)).fillna(0.0)

    return df

train_merged = add_topic_overlap_features(train_merged, topic_cols)
test_merged  = add_topic_overlap_features(test_merged, topic_cols)

```
