# DigiCow Farmer Training Adoption Challenge

## Exploratory Data Analysis

```python
# import libraries 

import pandas as pd 
import numpy as np 
```

```python
# load datasets

train = pd.read_csv("D:/digicow-farmer-training-adoption-challenge20260210-32396-19ubatl (1)/Train.csv")
test = pd.read_csv("D:/digicow-farmer-training-adoption-challenge20260210-32396-19ubatl (1)/Test.csv")
prior = pd.read_csv("D:/digicow-farmer-training-adoption-challenge20260210-32396-19ubatl (1)/Prior.csv")

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("Prior shape:", prior.shape)
```

```python
# view columns and first rows

print("Train columns:\n", train.columns.tolist())
print("\nTrain head:")
display(train.head())

print("Prior columns:\n", prior.columns.tolist())
print("\nPrior head:\n")
display(prior.head())
```

```python
# Clean column names (strip spaces)

def clean_columns(df):
    df.columns = df.columns.str.strip()
    return df

train = clean_columns(train)
test = clean_columns(test)
prior = clean_columns(prior)

print("Cleaned train columns:\n", train.columns.tolist())
print("Cleaned test columns:\n", test.columns.tolist())
print("Cleaned prior columns:\n", prior.columns.tolist())
```

```python
# Parse training_day as a date

for df, name in [(train, "train"), (test, "test"), (prior, "prior")]:
    if "training_day" in df.columns:
        df["training_day"] = pd.to_datetime(df["training_day"], errors = "coerce")
        print(name, "training_day parsed. Missing dates: ", df["training_day"].isna().sum())
    else:
        print(name, "does not have training_day column")   
```

```python
# Target monotonicity check (7 ⟶ 90 ⟶ 120)
# Logical rule: If adopted_within_07_days == 1, then adopted_within_90_days and adopted_within_120_days should also be 1.

t7 = "adopted_within_07_days"
t90 = "adopted_within_90_days"
t120 = "adopted_within_120_days"

if all(col in train.columns for col in [t7, t90, t120]):
    violations_7_90 = ((train[t7] == 1) & (train[t90] == 0)).sum()
    violations_7_120 = ((train[t7] == 1) & (train[t120] == 0)).sum()
    violations_90_120 = ((train[t90] == 1) & (train[t120] == 0)).sum()

    print("violations (7->90) = ", violations_7_90)
    print("violations (7->120) = ", violations_7_120)
    print("violations (90->120) = ", violations_90_120)

else:
    print("Target column names not found. Check train.columns")
```

```python
# Missing value report (overall + top missing columns)

missing = train.isna().mean().sort_values(ascending=False)*100
missing = missing[missing > 0]


print("Columns with missing values (%)")
display(missing.to_frame("Missing_%").head(20))
```

```python
# Cardinality check (how many unique values)
# High-cardinality categoricals can overfit. Check key columns.

cols_to_check = ['farmer_name', 'gender', 'registration', 'group_name', 'belong_to_cooperative', 'county', 'subcounty', 'ward', 'trainer', 'age']

for col in cols_to_check:
    if col in train.columns:
        print(f"{col:15s} unique values:", train[col].nunique(dropna=False))
```

```python
# Class imbalance (positive rates for each target)

for col in [t7, t90, t120]:
    if col in train.columns:
        pos_rate = train[col].mean()*100
        pos_count = train[col].sum()
        print(f"{col}: {pos_rate:.2f}% positives ({int(pos_count)} positive cases)")
```

```python
# Check if Prior actually contains multiple events per farmer

if "farmer_name" in prior.columns:
    events_per_farmer = prior.groupby("farmer_name").size()
    print("Prior: farmers:", events_per_farmer.shape[0])
    print("Prior: average events/farmer:", events_per_farmer.mean())
    print("Prior: max events for one farmer:", events_per_farmer.max())
    display(events_per_farmer.describe())
else:
    print("Prior does not contain farmer name column.")
```
